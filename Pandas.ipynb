#TITANIC DATASET ANALYSIS

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/Titanic-Dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

Missing values from Age, Cabin, and Embarked.


```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

Fill  Age missing values with median grouped by Pclass and Sex.

```python
df['Age'] = df.groupby(['Pclass', 'Sex'])['Age'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

Fill Missing Cabin Values with Unknown

```python
df['Cabin'] = df['Cabin'].str[0].fillna('Unknown')
```

```python
df.isnull().sum()
```

Fill missing Embarked with the modal value

```python
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

We can extract New Features

We can get family size by getting the count associated to a ticket

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.head()
```

We can also check if the Passenger is alone and get that into a feature

```python
df['IsAlone'] = (df['FamilySize'] == 1).astype(int)
```

```python
def get_vat(data: pd.Series, vat_rate:float = 0.75 ):
    """This Function accepts a series and returns the VAT calculated as the series.
    @param data(pd.Series): This is the Series for which the VAT is to be calculated
    @param vat_rate(float) : This is the rate of the VAT, usually between 0 and 1.
    @return Panda Series(pd.Series): It returns a Series with the Vat calculated."""
    return data * vat_rate

```

```python
df['VAT'] = df['Fare'].apply(get_vat)
get_vat(df['Fare'])
```
